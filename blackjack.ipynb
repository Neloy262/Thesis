```
import sys
import gym
import numpy as np
import random
from collections import defaultdict
env = gym.make('Blackjack-v0')
```

```
def generate_episodes(env):
  state_dict=dict()
  episodes=[]
  state=env.reset()
  while True:
    if state[0]>18:
      prob=[0.8,0.2]
    else:
      prob=[0.2,0.8]
    action=np.random.choice(np.arange(2),p=prob)
    next_state,reward,done,info=env.step(action)
    episodes.append([state,action,reward])
    state_dict[(state,action)]=reward
    state=next_state
    if done:
      break
  return episodes
```

```
def prediction(env,ep_num):
  state_action_value=dict()
  Q=dict()
  N=dict()
  R=0
  gamma=1
  for i in range(ep_num):
    ep=generate_episodes(env)
    G=0
    for i, e in reversed(list(enumerate(ep))):
      G=G*gamma+e[2]
      val=N.get((e[0],e[1]))
      if val==None:
        val=1
        N[(e[0],e[1])]=val
      else:
        val+=1
        N[(e[0],e[1])]=val
      state_val=state_action_value.get((e[0],e[1]))
      if state_val==None:
        state_action_value[(e[0],e[1])]=G
      else:
        state_val+=G
        state_action_value[(e[0],e[1])]=state_val
  
  key_list=state_action_value.keys()

  for i in key_list:
    state_action_value[i]=state_action_value[i]/N[i]
      
  return state_action_value
```

```
Q_val=prediction(env,10000)
for i in Q_val:
  print("State:",i,"Value:",Q_val.get(i))
```
